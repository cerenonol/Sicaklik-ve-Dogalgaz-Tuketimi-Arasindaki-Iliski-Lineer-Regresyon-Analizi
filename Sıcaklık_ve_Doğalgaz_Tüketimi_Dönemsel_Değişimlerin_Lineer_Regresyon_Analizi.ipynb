```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data={
    'sehir': ['Eskisehir', 'İstanbul', 'Bursa', 'Nevşehir', 'Diyarbakır',
              'Kütahya', 'Mersin', 'Uşak', 'Kırşehir', 'Bartın', 'Giresun',
               'Sinop', 'Ordu'],
    'yaz_sicaklik': [24.3,26.7,28,28.3,34.3,
                     26,29,27,27,24.7,23.7,
                     25.7,24.7],
    'kis_sicaklik': [4.3,8.7,7.3,2.3,5.3,3.3,
                     11,4.3,2.3,6.3,7.3,
                     7.3,6.3],
    'dogalgaz_tuketimi': [15,27.60,10.20,47.70,10.40,20.10,
                          22,22.10,14.10,12.80,18.30,26.30,
                          22.60]
      }

df=pd.DataFrame(data)

print("DataFrame sütunları:", df.columns)

X_yaz = df [['yaz_sicaklik']]
y_yaz = df [['dogalgaz_tuketimi']]
X_kis = df [['kis_sicaklik']]

model_yaz = LinearRegression()
model_yaz.fit(X_yaz, y_yaz)

y_pred_yaz=model_yaz.predict(X_yaz)

plt.figure(figsize=(12,6))
plt.scatter(df['yaz_sicaklik'],df['dogalgaz_tuketimi'], color='blue', label='Veri Noktaları')
plt.plot(df['yaz_sicaklik'],y_pred_yaz, color='red', label='Regresyon Doğrusu')
plt.title('Yaz Sıcaklıkları ve Doğalgaz Tüketimi')
plt.xlabel('Yaz Sıcaklık Ortalaması (C)')
plt.ylabel('Doğalgaz Tüketimi (%)')
plt.legend()
plt.grid()
plt.show()


X_kis =df [['kis_sicaklik']]
y_kis =df [['dogalgaz_tuketimi']]

model_kis = LinearRegression()
model_kis.fit(X_kis, y_kis)

y_pred_kis=model_kis.predict(X_kis)

plt.figure(figsize=(12,6))
plt.scatter(df['kis_sicaklik'],df['dogalgaz_tuketimi'], color='green', label='Veri Noktaları')
plt.plot(df['kis_sicaklik'],y_pred_kis, color='red', label='Regresyon Doğrusu')
plt.title('Kış Sıcaklıkları ve Doğalgaz Tüketimi')
plt.xlabel('Kış Sıcaklık Ortalaması (C)')
plt.ylabel('Doğalgaz Tüketimi (%)')
plt.legend()
plt.grid()
plt.show()

```
